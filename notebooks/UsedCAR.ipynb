```python
num=x_train.select_dtypes(include=['int64','float64']).columns
cat=x_train.select_dtypes(include=['object']).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
num_pipeline=Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scaler',StandardScaler())
])
```

```python
# Handle unknown categories by setting handle_unknown='ignore' in OneHotEncoder
cat_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))  # Added handle_unknown='ignore'
])
```

```python
# Import the necessary class
from sklearn.compose import ColumnTransformer

# Rest of your code remains the same:
preprocessor = ColumnTransformer([
    ('num_pipeline', num_pipeline, num),
    ('cat_pipeline', cat_pipeline, cat)
])
```

```python
preprocessor
```

```python
x_train=preprocessor.fit_transform(x_train)
x_test=preprocessor.transform(x_test)
```
